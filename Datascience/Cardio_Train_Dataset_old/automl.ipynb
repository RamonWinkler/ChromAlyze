```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tpot import TPOTClassifier

# Drop rows with missing values or handle them accordingly
df_clean = df.dropna()

# Define features (X) and target (y)
X = df_clean.drop(["cardio"], axis=1)
y = df_clean["cardio"]

# Split the data into training, validation, and test sets
train_ratio = 0.70
validation_ratio = 0.15
test_ratio = 0.15

# First split into train and test (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1 - train_ratio, random_state=42)

# Then split the remaining 30% into validation and test (15% validation, 15% test)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=test_ratio / (test_ratio + validation_ratio), random_state=42)

# Initialize and fit TPOTClassifier (with verbosity to see progress)
pipeline_optimizer = TPOTClassifier(generations=5, population_size=20, cv=5, verbosity=2, random_state=42)

# Fit the model to the training data
pipeline_optimizer.fit(X_train, y_train)

# Evaluate the model on the test set
print("Test accuracy: ", pipeline_optimizer.score(X_test, y_test))
```
